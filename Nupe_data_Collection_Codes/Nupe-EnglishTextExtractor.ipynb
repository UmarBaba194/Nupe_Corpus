```python
import re
import pandas as pd

# Phase 2: EnglishTextTokenizer
def english_text_tokenizer(english_text):
    # Use a more robust sentence tokenizer
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', english_text)
    return sentences

# Phase 3: EnglishTokenSaver
def save_english_tokens_to_csv(tokens, csv_path):
    df = pd.DataFrame({"English_text": tokens})
    df.to_csv(csv_path, index=False)

# Phase 5: NupeTextTokenizer
def nupe_text_tokenizer(nupe_text):
    # Use a more robust sentence tokenizer
    sentences = re.split(r'(?<!\w\.\w.)(?<![A-Z][a-z]\.)(?<=\.|\?|\!)\s', nupe_text)
    return sentences

# Phase 6: NupeTokenSaver
def save_nupe_tokens_to_csv(tokens, csv_path):
    df = pd.DataFrame({"Nupe_text": tokens})
    df.to_csv(csv_path, index=False)

# Phase 7: Merger
def merge_records(csv_path_english, csv_path_nupe, merged_csv_path):
    df_english = pd.read_csv(csv_path_english)
    df_nupe = pd.read_csv(csv_path_nupe)

    # Merging records on index
    merged_df = pd.merge(df_english, df_nupe, left_index=True, right_index=True)
    merged_df.to_csv(merged_csv_path, index=False)

# Example usage:
english_text = "This is an example English sentence. Another sentence follows."
nupe_text = "Enyin yaya Igala ta?. Ai Ebi da agaba."

# Tokenizing English text
english_tokens = english_text_tokenizer(english_text)
# Saving English tokens to CSV
save_english_tokens_to_csv(english_tokens, "english_tokens.csv")

# Tokenizing Nupe text
nupe_tokens = nupe_text_tokenizer(nupe_text)
# Saving Nupe tokens to CSV
save_nupe_tokens_to_csv(nupe_tokens, "nupe_tokens.csv")

# Merging records
merge_records("english_tokens.csv", "nupe_tokens.csv", "merged_records.csv")

# Read and print the merged records CSV
merged_df = pd.read_csv("merged_records.csv")
print(merged_df)
```
