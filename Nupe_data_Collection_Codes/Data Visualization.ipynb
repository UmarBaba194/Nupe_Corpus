```python
import pandas as pd
```

```python
df=pd.read_csv('nupe-engFinal2.csv')
```

```python
df['nupe words count']=df['nupe'].str.split().str.len()
```

```python
df['eng words count']=df['eng'].str.split().str.len()
```

```python
df.head()
```

```python
df['eng words count'].max()
```

```python
df['nupe words count'].max()
```

```python
df=df[(df['eng words count']<320)&(df['nupe words count']<320)]
```

```python
df
```

```python
# Unique Words in Nupe
nupe_unique_words = set(df['nupe'].str.split().explode())
print("Unique Words in Nupe:", len(nupe_unique_words))

# Total Number of Unique Words
total_unique_words = len(set(df['nupe'].str.split().explode()))
print("Total Number of Unique Words:", total_unique_words)
```

```python
from sklearn.model_selection import train_test_split
train,test=train_test_split(df,test_size=0.2)
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Concatenate all Nupe words into a single string
nupe_text = ' '.join(df['nupe'].astype(str))

# Generate a WordCloud object
wordcloud = WordCloud(width=800, height=400, background_color='black').generate(nupe_text)

# Plot the WordCloud image
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.title('Word Cloud for Nupe Words')
plt.show()
```
